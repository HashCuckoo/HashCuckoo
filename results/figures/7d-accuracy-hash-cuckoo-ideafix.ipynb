# Hash Analysis - HashCuckoo and IDEAFIX Accuracy

#### Figure 7(d)

```python
#-*- coding: UTF-8 -*-
"""
========
Hash Analysis - HashCuckoo and IDAFIX Accuracy
========

"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
import pylab as pyl
#import pickle as p
from matplotlib import rc
#import pickle as p
# rc('font',**{'family':'serif','serif':['Times']})
# rc('text', usetex=True)

N = 6
SIZE=16

FP_HC      = [87.248, 73.971, 61.452, 55.031, 52.012, 50.201]   #HashCuckoo S2 - False Positives
FN_HC      = [75.320, 63.240, 51.180, 48.201, 47.123, 46.010]   #HashCuckoo S2 - False Negatives
FP_ideafix = [33.340, 22.480, 14.530,  8.600,  3.770,  1.810]   #IDEAFIX - False Positives
FN_ideafix = [ 5.248,  3.197,  1.459,  0.243,  0.102,  0.095]   #IDEAFIX - False Negatives
ind = np.arange(N)  # the x locations for the groups
width = 0.6    # the width of the bars

fig, ax = plt.subplots(figsize=(7, 4))

plt.plot(ind, FP_HC, color='#388E3C', marker='^', linestyle='--', label='HashCuckoo - False Positives')
plt.plot(ind, FN_HC, color='#69f0ae', marker='o', linestyle='--', label='HashCuckoo - False Negatives')

plt.plot(ind, FP_ideafix, color='#007c91', marker='s', linestyle='--', label='IDEAFIX - False Positives')
plt.plot(ind, FN_ideafix, color='#4dd0e1', marker='x', linestyle='--', label='IDEAFIX - False Negatives')


# add some text for labels, title and axes ticks
plt.xticks(fontsize=SIZE)
plt.yticks(fontsize=SIZE)

ax.set_ylabel('Flows Identified (%)' , fontsize=SIZE)
ax.set_xlabel('Memory Space for Hash Mapping (%)' , fontsize=SIZE)
ax.set_xticks([0,1,2,3,4,5])
ax.set_xticklabels(('25%', '50%', '75%', '100%', '125%', '150%'), fontsize=SIZE, rotation=0)
ax.set_yticks(np.arange(0, 101,20))
ax.legend(fontsize=SIZE-6, ncol=2,loc=1, edgecolor='black')


ax.set_axisbelow(True)
ax.yaxis.grid(color='silver', linestyle='-')
ax.xaxis.grid(color='silver', linestyle='-')

ax.tick_params(direction='inout', length=10)
ax.tick_params( which='minor', direction='out', length=5)

fig.tight_layout()

plt.show()
fig.savefig("fig/accuracy-hashCuckoo-Ideafix.pdf")
```
