```python
#-*- coding: UTF-8 -*-
"""
========
Hash Analysis - HashCuckoo Accuracy
========

"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
import pylab as pyl
#import pickle as p
from matplotlib import rc
#import pickle as p
# rc('font',**{'family':'serif','serif':['Times']})
# rc('text', usetex=True)



N = 6
SIZE=16

FP_s2 = [87.248, 73.971, 61.452, 55.031, 52.012, 50.201]   #S2 - False Positives
FN_s2 = [75.320, 63.240, 51.180, 48.201, 47.123, 46.010]   #S2 - False Negatives
FP_s1 = [75.340, 58.480, 46.530, 42.600, 41.770, 40.810]   #S1 - False Positives
FN_s1 = [63.248, 48.197, 40.459, 38.243, 37.000, 36.600]   #S1 - False Negatives

ind = np.arange(N)  # the x locations for the groups
width = 0.6    # the width of the bars

fig, ax = plt.subplots(figsize=(7, 4))

plt.plot(ind, FP_s2, color='#255d00', marker='^', linestyle='--', label='S2 - False Positives')
plt.plot(ind, FN_s2, color='#69f0ae', marker='o', linestyle='--', label='S2 - False Negatives')

plt.plot(ind, FP_s1, color='r',       marker='s', linestyle='--', label='S1 - False Positives')
plt.plot(ind, FN_s1, color='#ff9800', marker='x', linestyle='--', label='S1 - False Negatives')



#S2 - False Positives text
ax.text(3.25, (FP_s2[3]),''.join(str(int(FP_s2[3]))+"%"),
                ha='center', va='bottom', color='#255d00', fontsize=SIZE-6)

#S1 - False Negatives
ax.text(3.25, (FN_s1[3]-10.5),''.join(str(int(FN_s1[3]))+"%"),
                ha='center', va='bottom', color='#ff9800', fontsize=SIZE-6)


# Add some text for labels, title and axes ticks
plt.xticks(fontsize=SIZE)
plt.yticks(fontsize=SIZE)
ax.set_ylabel('Flows Identified (%)' , fontsize=SIZE)
ax.set_xlabel('Memory Space for Hash Mapping (%)' , fontsize=SIZE)
ax.set_xticks([0,1,2,3,4,5])
ax.set_xticklabels(('25%', '50%', '75%', '100%', '125%', '150%'), fontsize=SIZE, rotation=0)
ax.set_yticks(np.arange(0, 101,20))
ax.legend(fontsize=SIZE-6, ncol=2,loc=1, edgecolor='black')


ax.set_axisbelow(True)
ax.yaxis.grid(color='silver', linestyle='-')
ax.xaxis.grid(color='silver', linestyle='-')


ax.tick_params(direction='inout', length=10)
ax.tick_params( which='minor', direction='out', length=5)


fig.tight_layout()
plt.show()
fig.savefig("fig/accuracyHash.pdf")
```
